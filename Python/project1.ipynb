# Data Preprocessing And Preparation

```python
#keep a copy of the data before preprocessing
```

```python
df_mpi_2 = df_mpi.copy()
df_mpi_2.head()
```

```python
#checking for missing values
```

```python
df_mpi_2.isnull().head()
```

```python
df_mpi_2.isnull().sum()
```

```python
#replace missing values with a number. Note that the missing values are eight and they are found in a singular column.
```

```python
df_mpi_2['MPI of the country'] = df_mpi_2['MPI of the country'].replace(np.nan, 0.25)
df_mpi_2.head()
```

```python
df_mpi_2.isnull().sum()
```

```python
#There are no more missing values as the isnull.sum is now 0
```

```python
#drop columns not needed for the analysis
```

```python
df_mpi_2.drop(['ISO Country Code', 'Country', 'World region'], axis = 1)
df_mpi_2 = df_mpi_2.drop(['ISO Country Code', 'Country', 'World region'], axis = 1)
```

```python
df_mpi_2.head()
```

```python
#Before proceeding to data visualization, let us return the rows of the states with high Multidimensional Poverty Index - MPI
```

```python
df_mpi_2.loc[df_mpi_2['MPI By States = %H * %A'] >= 0.35]
df_mpi_2 = df_mpi_2.loc[df_mpi_2['MPI By States = %H * %A'] >= 0.35]
df_mpi_2
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (9, 6))
plt.bar(x = df_mpi_2["State"],
       height = df_mpi_2["No Of MPI In Each State"])
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Multidimensional Poverty Index In Nothern Nigeria", fontsize = 16, fontweight = "bold")
plt.ylabel("No of MPI In Millions", fontsize = 13)
plt.show()
plt.savefig("MPI-In-Nothern-Nigeria.png")
```

# Regression Analysis

```python
plt.figure(figsize = (10, 8))
sns.regplot(x = 'State Population Size',
           y = 'No Of MPI In Each State',
           data = df_mpi_2,
           scatter_kws = {'color': 'k'},
           line_kws = {'color': 'red'})
plt.xlabel("Population of States in Millions")
plt.ylabel("No Of MPI in Millions")
plt.title("Effect Of Population size on Multidimensional Poverty Index", fontsize = 14, weight = "bold")
plt.show()
```
